### 0. Taylor Expansion
$f(x) = \sum_{k=1}^{\infty}\frac{f^{(k)}(a)}{k!}(x-a)^{k}$

이때 차수가 높아질수록, a에 가까워 질수록 더 정확하게 흉내낼 수 있다.

ex) $f(x) = 20\sin(2x)+ exp(2x)$

```python
def f(x):
    return 20*np.sin(x*2)+np.exp(x*2)
```

```python
x_range = [(i-500)/250 for i in range(1000)]
plt.figure(figsize=(7,5))
plt.plot(x_range,[f(x) for x in x_range])
plt.show()
```

```python
def fp1(x):
    return 20*np.cos(x*2)*2+2*np.exp(x*2)

a = -0.5

f1 = [f(a) + fp1(a)*(i-a) for i in x_range]
```

```python
x_range = [(i-500)/250 for i in range(1000)]
plt.figure(figsize=(7,5))
plt.plot(x_range,[f(x) for x in x_range])
plt.plot(x_range,f1,color="red")
plt.ylim(-30,40)
plt.show()
```

```python
def fp1(x):
    return 20*np.cos(x*2)*2+2*np.exp(x*2)
def fp2(x):
    return -20*np.sin(x*2)*2**(2)+2**(2)*np.exp(x*2)
def fp3(x):
    return 20*np.cos(x*2)*2**(3)+2**(3)*np.exp(x*2)
def fp4(x):
    return -20*np.sin(x*2)*2**(4)+2**(4)*np.exp(x*2)
def fp5(x):
    return 20*np.cos(x*2)*2**(5)+2**(5)*np.exp(x*2)

a = -0.5

f1 = [f(a) + fp1(a)*(i-a) for i in x_range]
f2 = [f(a) + fp1(a)*(i-a)+ (fp2(a)/2)*(i-a)**2 for i in x_range]
f3 = [f(a) + fp1(a)*(i-a)+ (fp2(a)/2)*(i-a)**2 + (fp3(a)/3/2)*(i-a)**3 for i in x_range]
f4 = [f(a) + fp1(a)*(i-a)+ (fp2(a)/2)*(i-a)**2 + (fp3(a)/3/2)*(i-a)**3 + (fp4(a)/4/3/2)*(i-a)**4  for i in x_range]
f5 = [f(a) + fp1(a)*(i-a)+ (fp2(a)/2)*(i-a)**2 + (fp3(a)/3/2)*(i-a)**3 + (fp4(a)/4/3/2)*(i-a)**4 + (fp5(a)/5/4/3/2)*(i-a)**5/5/4/3/2 for i in x_range]

```

```python
x_range = [(i-500)/250 for i in range(1000)]
plt.figure(figsize=(14,10))
plt.plot(x_range,[f(x) for x in x_range],linewidth=4)
plt.plot(x_range,f1,linestyle='dashed',label = "k=1")
plt.plot(x_range,f2,linestyle='dashed',label = "k=2")
plt.plot(x_range,f3,linestyle='dashed',label = "k=3")
plt.plot(x_range,f4,linestyle='dashed',label = "k=4")
plt.plot(x_range,f5,linestyle='dashed',label = "k=5")
plt.legend()
plt.ylim(-30,40)
plt.show()
```

### 1. Gradient Descent

For f:diff,convex, a sequence $x_n = x_{n-1}-t \nabla f(x_{t-1})$ converges to $\underset{x}{\arg \min} f(x)$ for appropriate t.

A. $f(x) = x^2 + x + 1$  

$x_n = x_{n-1} - t f'(x_{n-1}) = x_{n-1}-t (2x_{n-1}+1)$

```python
def f(x) :
    return x**2+x+1
def nablaf(x):
    return 2*x+1
```

```python
x_init = 0
stepsize = 0.001

x_temp = x_init
for i in range(5000):
    x_temp += -1*stepsize*nablaf(x_temp)
```

```python
x_range = [(i-500)/250 for i in range(1000)]
plt.figure(figsize=(7,5))
plt.plot(x_range,[f(x) for x in x_range])
plt.vlines(x_temp,ymin=0,ymax=8,linestyles="dashed",color="red")
plt.show()
```

### c. Linear Regression
$y = X\beta$  
$f(\beta) = (y-X\beta)^t(y-X\beta)$

```python
x0 = np.ones((1000))
x1 = np.random.normal(0,1,1000)
x2 = np.random.normal(0,1,1000)
x3 = np.random.normal(0,1,1000)
X = np.concatenate([[x0],[x1],[x2],[x3]],axis=0).T
beta = np.array([1,5,4,3]).reshape((-1,1))
y= X@beta+np.random.normal(0,1,1000).reshape((-1,1))
```

```python
plt.figure(figsize=(10,4))
plt.subplot(131)
plt.plot(X[:,1],y,"o")
plt.subplot(132)
plt.plot(X[:,2],y,"o")
plt.subplot(133)
plt.plot(X[:,3],y,"o")
plt.show()
```

```python
def f(beta,X,y):
    return (y-X@beta).T@(y-X@beta)
def nablaf(beta,X,y):
    return -2*X.T@y+2*X.T@X@beta
```

```python
beta_init = np.array([1.0,1.0,1.0,1.0]).reshape(-1,1)
stepsize = 0.0001

beta_temp = beta_init
for i in range(10000):
    beta_temp += -stepsize*nablaf(beta_temp,X,y)
```

### 4. Accerlating Methods

#### Backtracking Line Search

```python
beta_init = np.array([1.0,1.0,1.0,1.0]).reshape(-1,1)
stepsize = 1
n_iter= 10
alpha = 0.5
beta = 0.5 

beta_list = []
step_list = []
beta_temp = beta_init

for i in range(n_iter):
    gradient = nablaf(beta_temp,X,y)
    while gradient.T@gradient*alpha*stepsize > f(beta_temp,X,y)-f(beta_temp-stepsize*nablaf(beta_temp,X,y),X,y):
        stepsize = beta*stepsize 
    step_list += [stepsize]
    beta_temp = beta_temp-stepsize*nablaf(beta_temp,X,y)
    beta_list += [beta_temp]
plt.plot(range(n_iter), [f(beta,X,y)[0][0] for beta in beta_list])
```

#### Strong Covexity Search

```python
L = np.max(np.linalg.eigh(2*X.T@X)[0])
m = np.min(np.linalg.eigh(2*X.T@X)[0])
str_con = 2/(L+m)
```

### 5. Stopping Criterion

```python
beta_init = np.array([1.0,1.0,1.0,1.0]).reshape(-1,1)
stepsize = 0.0000001
n_iter= 100000
epsilon = 0.001

beta_list = []
beta_temp = beta_init

for i in range(n_iter):
    beta_temp = beta_temp-stepsize*nablaf(beta_temp,X,y)
    beta_list += [beta_temp]
    gradient = nablaf(beta_temp,X,y)
    if gradient.T@gradient < np.sqrt(2*m*epsilon) : break

plt.plot(range(i+1), [f(beta,X,y)[0][0] for beta in beta_list])
```
